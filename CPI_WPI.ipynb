```python
# accumulated percentage Private
Private_percent_increase = percent_df['Percentage Change from Previous Quarter Total hourly rates of pay excluding bonuses Australia Private All industries'].sum()
print(f"From 1/12/1997 There is an accumulated percentage increase in wages in the Private sector by: {Private_percent_increase}%" )
Private_price_increase = df['Total hourly rates of pay Private'].max() - df['Total hourly rates of pay Private'].min()
min = df['Total hourly rates of pay Private'].min()
max = df['Total hourly rates of pay Private'].max()
print(f"The difference in Private wage price increase is ${Private_price_increase}")
print(f"The min is ${min}")
print(f"The max is ${max}")
```

```python
# accumulated percent public
Public_percent_increase = percent_df['Percentage Change from Previous Quarter Total hourly rates of pay excluding bonuses Australia Public All industries '].sum()
print(f"From 1/12/1997 There is an accumulated percentage increase in wages in the Public sector by: {Public_percent_increase}%" )
Public_price_increase = df['Total hourly rates of pay Public '].max() - df['Total hourly rates of pay Public '].min()
min = df['Total hourly rates of pay Public '].min()
max = df['Total hourly rates of pay Public '].max()
print(f"The difference in Public wage price increase is ${Public_price_increase}")
print(f"The min is ${min}")
print(f"The max is ${max}")
```
